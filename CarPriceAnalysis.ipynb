# Which car brand has the highest average price?

```python
df['Brand'] = df['CarName'].str.split().str[0]
df.groupby('Brand')['price'].mean().sort_values(ascending=False).head(10)
```

# Is horsepower correlated with car price?

```python
df[['horsepower', 'price']].corr()
```

# What fuel type (gas vs diesel) has higher average price?

```python
df.groupby('fueltype')['price'].mean()
```

# What is the average mileage (city & highway mpg) per car body type?

```python
df.groupby('carbody')[['citympg', 'highwaympg']].mean()
```

# Which 5 cars are the most expensive in the dataset?

```python
df[['CarName', 'price']].sort_values('price', ascending=False).head(5)
```

# Horsepower vs Price

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x='horsepower', y='price', hue='fueltype')
plt.title("Horsepower vs Price")
plt.show()
```

# Price Distribution by Car Body Type

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='carbody', y='price')
plt.title("Price Distribution by Car Body Type")
plt.xticks(rotation=45)
plt.show()
```

# Feature Correlation Matrix

```python
plt.figure(figsize=(12,8))
sns.heatmap(df[['price','horsepower','enginesize','citympg','highwaympg','compressionratio']].corr(),
            annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

# Sales by Category 

```python
categories = ["Electronics", "Furniture", "Clothing", "Grocery"]
sales = [120000, 85000, 60000, 150000]

plt.figure(figsize=(8,5))
plt.bar(categories, sales)
plt.title("Sales by Category")
plt.xlabel("Category")
plt.ylabel("Sales ($)")
plt.tight_layout()
plt.show()
```

# Average Price by Brand

```python
brands = ["Brand A", "Brand B", "Brand C", "Brand D"]
avg_price = [45, 60, 35, 80]

plt.figure(figsize=(8,5))
plt.bar(brands, avg_price)
plt.title("Average Product Price by Brand")
plt.xlabel("Brand")
plt.ylabel("Average Price ($)")
plt.tight_layout()
plt.show()
```
